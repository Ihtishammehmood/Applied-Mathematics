P 5.35)

The distance travelled by a body falling freely in space is described
by the equation
$$

$$
, where d is in feet, and t is in seconds. Draw a
curve showing the relation between d and t. Determine the velocity of the
body at the following times: t = 2[s], t = 4.6[s], and t = 0.01[s].

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variables
t = sp.Symbol('t')

# Define the distance equation
d = 16*t**2

# Take the derivative to find the velocity
v = sp.diff(d, t)

# Evaluate the velocity at the given times
seconds = [2, 4.6, 0.01]
for s in seconds:
    rate = v.evalf(subs={t: s})
    print(f"Velocity of body at {s} seconds is {rate:.2f} ft/s")


t_vals = np.linspace(0,5,100)

# compute d and v values of  t values
d_vals = [d.subs(t,vals)for vals in t_vals]
v_vals = [v.subs(t,vals)for vals in t_vals]

# Distance-time curve
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(t_vals,d_vals)
plt.xlabel('Time(seconds)')
plt.ylabel('Distance(Feet)')
plt.title('Distance-Time Curve')
plt.grid()

# Velocity-time curve

plt.subplot(1,2,2)
plt.plot(t_vals,v_vals)
plt.xlabel('Time (Seconds)')
plt.ylabel('Velocity (feet/second)')
plt.title('Velocity-time-curve')
plt.grid()

plt.tight_layout()
plt.show()
```

$$
where n is a constant. Find the value of n when the velocity is doubled
from the 5th to the 10th second. Find also the value of n when the velocity

$$

**Saddle Points**:Saddle points in a multivariable function are those critical points where the function attains neither a local maximum value nor a local minimum value.

```python
import sympy as sp

x = sp.symbols('x')
y = 1- (x**2/2)+(x**4/24)

dy_dx = sp.diff(y,x)
critical_points=sp.solve(dy_dx,x)
d2y_dx2 = sp.diff(dy_dx,x)
# initialize the variables to count maxima and minima
maxima_count = 0
minima_count = 0
# check the 2nd derivative at each critical point to determine wether its minimum or maximum
for point in critical_points:
  point = sp.N(point) # converting to floating point number
  second_derivative_value = d2y_dx2.subs(x,point)
  if second_derivative_value >0:
    minima_count +=1
  elif second_derivative_value <0:
    maxima_count +=1
print(f"Number of maxima:{maxima_count}")
print(f"Number of minima:{minima_count}")
```
